```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate a sample dataset
np.random.seed(42)  # For reproducibility
n = 1000  # Number of samples

# Generate age data
ages = np.random.randint(0, 100, size=n)

# Generate gender data
genders = np.random.choice(['Male', 'Female'], size=n)

# Create a DataFrame
sample_df = pd.DataFrame({'Age': ages, 'Gender': genders})

# Display the first few rows of the sample dataset
print(sample_df.head())

# Bar chart for gender distribution
gender_counts = sample_df['Gender'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(gender_counts.index, gender_counts.values, color=['skyblue', 'lightcoral'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.show()

# Histogram for age distribution
plt.figure(figsize=(10, 6))
plt.hist(sample_df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```
