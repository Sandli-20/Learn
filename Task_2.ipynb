```python
import pandas as pd

# Create a small sample dataset
data = {
    'PassengerId': [1, 2, 3, 4, 5],
    'Pclass': [3, 1, 3, 1, 3],
    'Name': [
        'Braund, Mr. Owen Harris',
        'Cumings, Mrs. John Bradley (Florence Briggs Thayer)',
        'Heikkinen, Miss. Laina',
        'Futrelle, Mrs. Jacques Heath (Lily May Peel)',
        'Allen, Mr. William Henry'
    ],
    'Sex': ['male', 'female', 'female', 'female', 'male'],
    'Age': [22, 38, 26, 35, 35],
    'Fare': [7.25, 71.2833, 7.925, 53.1, 8.05],
    'Embarked': ['S', 'C', 'S', 'S', 'S'],
    'Survived': [0, 1, 1, 1, 0]
}

# Convert the dictionary to a DataFrame
sample_data = pd.DataFrame(data)

# Display the sample dataset
print(sample_data)
```

```python
# Check for missing values in the sample dataset
print(sample_data.isnull().sum())

# Convert 'Sex' and 'Embarked' into numerical values
sample_data['Sex'] = sample_data['Sex'].map({'male': 0, 'female': 1})
sample_data['Embarked'] = sample_data['Embarked'].map({'S': 0, 'C': 1})

print(sample_data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the distribution of 'Age'
plt.figure(figsize=(8, 5))
sns.histplot(sample_data['Age'], bins=5, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the distribution of 'Fare'
plt.figure(figsize=(8, 5))
sns.histplot(sample_data['Fare'], bins=5, kde=True)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot survival rate by 'Pclass'
plt.figure(figsize=(8, 5))
sns.barplot(x='Pclass', y='Survived', data=sample_data)
plt.title('Survival Rate by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Age vs. Fare colored by 'Survived'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=sample_data, palette='coolwarm', style='Survived', s=100)
plt.title('Age vs. Fare by Survival Status')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
