```python
import random
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Generate random social media data
def generate_social_media_data(num_posts):
    data = []
    for i in range(num_posts):
        post_id = i + 1
        user_id = random.randint(1, 100)
        post_content = f"This is a synthetic post number {post_id}"
        timestamp = datetime.now() - timedelta(days=random.randint(0, 30))
        sentiment_score = round(random.uniform(-1, 1), 2)
        data.append([post_id, user_id, post_content, timestamp, sentiment_score])
    return pd.DataFrame(data, columns=['Post ID', 'User ID', 'Content', 'Timestamp', 'Sentiment'])

# Generate 100 synthetic posts
df = generate_social_media_data(100)
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the visual style
sns.set(style="whitegrid")

# Sentiment Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['Sentiment'], bins=20, kde=True)
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.show()

# Time Series Analysis
df['Date'] = df['Timestamp'].dt.date
sentiment_over_time = df.groupby('Date')['Sentiment'].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(x='Date', y='Sentiment', data=sentiment_over_time)
plt.title('Average Sentiment Over Time')
plt.xlabel('Date')
plt.ylabel('Average Sentiment Score')
plt.show()

# User Sentiment Analysis
user_sentiment = df.groupby('User ID')['Sentiment'].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(x='User ID', y='Sentiment', data=user_sentiment)
plt.title('Average Sentiment per User')
plt.xlabel('User ID')
plt.ylabel('Average Sentiment Score')
plt.xticks(rotation=90)
plt.show()
```
